```python
feature_cols = ['country','province','winery','region_1'] #categorical feature cols
```

```python
len(df['winery'].unique())
```

```python
df= df.reset_index()
df
```

```python
df1 = pd.get_dummies(df[feature_cols])  #Columns with categorical variables without variety
```

```python
df1
```

```python
df1 = df1.reset_index()
```

```python
df1
```

```python
result = pd.concat([df,df1],axis=1)  #concat  categorical with encoded columns
```

```python
result.drop(feature_cols,axis=1,inplace=True)   #remove original categorical columns
```

```python
result = result.reset_index()
```

```python
result.drop('level_0',axis=1,inplace=True)
```

```python
result
```

```python
result.drop('index',axis=1,inplace=True)
```

```python
result = result.drop('designation',axis=1)
```

```python
result.shape
```

```python
X = result.drop('variety',axis=1)
y = result['variety']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train
```

# Model 1 : Decision Tree Classifier

```python
clf = DecisionTreeClassifier()   #Decision tree classifier
```

```python
clf = clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
pred = pd.DataFrame({'y_test':y_test,'y_pred':y_pred})
pred
```

# Model 2 : Random Forest Classifier

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf = rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_rf))
```

```python
pred_rf= pd.DataFrame({'y_test':y_test,'y_pred':y_pred_rf})
pred_rf
```
